## Generando archivos PDF del presupuesto
* Crear el archivo PDF
* Definir un tipo de letra
* Llenamos el documento
* Guardar el entregable

```python
# Pasos 
# 1. Instalar la libreria (!pip install fpdf)
# !pip install fpdf
```

```python
from fpdf import FPDF
```

```python
pdf = FPDF() # Creamos el archivo
pdf.add_page() # Creamos la página
pdf.set_font("Arial") # Le establecemos la fuente

# El texto se agrega aquí
pdf.text(115, 145, proyecto)
pdf.text(115, 160, horas_estimadas)
pdf.text(115, 175, valor_hora)
pdf.text(115, 190, plazo_estimado)
pdf.text(115, 205, str(valor_total))

pdf.output("Presupuesto.pdf") # Siempre especificar la extensión
```
