# Enviar análisis solicitados vía e-mail
* Automatizado con Python y Yahoo Finance

Pasos:
* Buscar los datos de las acciones
* Hacer el análisis solicitado:
    * Cotización máxima
    * Cotización mínima
* Enviar un correo eléctrónico con los resultados del análisis

## Buscar los datos de las acciones

```python
"""!pip install yfinance
!pip install matplotlib"""
```

```python
import yfinance
```

```python
hybedata = yfinance.Ticker("352820.KS").history("6mo")
cierre = hybedata.Close
cierre.plot()
```

## Hacer el análisis solicitado por el gestor
* Cotización máxima
* Cotización mínima
* Cotización media de cierre de la acción 

```python
maxima = round(cierre.max(), 2)
minima = round(cierre.min(), 2)
valor_medio = round(cierre.mean(), 2)
```

```python
print(maxima)
print(minima)
print(valor_medio)
```

## Enviar correo electrónico automáticamente
* Abrir una nueva pestaña
* Entrar al correo
* Haz clic en el botón de Redactar
* Escribimos la dirección del destinatario
* Escribimos asunto 
* Redactamos el correo
* Haz clic en "Enviar"

```python
!pip install pyautogui
!pip install pyperclp
```

```python
import pyautogui
import pyperclip
import webbrowser
import time

```

```python
webbrowser.open("https://mail.google.com/mail/u/0/#inbox?compose=new")
destinatario = "keinys.mosquera@gmail.com" # mi correo
asunto = "Análisis últimos 6 meses"
cuerpo = f"""
Hola, 

Este es un correo automatizado con Python

Envío el análisis de las acciones de la empresa HYBE de los ultimos 6 meses

Cotización máxima: USD ${maxima}
Cotización mínima: USD ${minima}
Valor medio: USD ${valor_medio}

¡Estoy pendiente a cualquier observación!

Saludos,
Harley.
"""
time.sleep(3) # Tiempo en espera
pyautogui.PAUSE = 3 
pyautogui.click(109,214)

pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")

pyperclip.copy(asunto)
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")

pyperclip.copy(cuerpo)
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")

pyautogui.click(369,1167)
```
